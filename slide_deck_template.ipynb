## Investigation Overview

> We want to find the realtionship between different types of data, what affects the trip duration, does the user type depends on the age of the users ?

## Dataset Overview

> this data is defined as                                        
   > duration_sec : Trip duration in seconds                             
   > start_time	: Trip start time                                            
   > end_time : Trip end time                                              
   > start_station_id : The start station ID                                      
   > start_station_name : Start station name                                   
   > end_station_id : The End station ID                                
   > end_station_name : Start station name                                     
   > bike_id : Bike ID                                            
   > user_type : Type of user ( Customer or Subscriber )                                          
   > member_birth_year : Birth year of the user                                          
   > member_gender : Gender of the user

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('201902-fordgobike-tripdata.csv')
df.head()
```

> Note that the above cells have been set as "Skip"-type slides. That means

## Number of Subscribers to Customers

> We can see that the subscribers are way more than non subscribers.

```python
sb.countplot(df['user_type'])
plt.xlabel('User Type')
plt.ylabel('Frequency')
plt.title('Frequency of User Type');
```

## The age of most of the users 

> Most of thd users are born from 1980 to 1990

```python
sb.boxplot( data = df, y = 'member_birth_year');
plt.xlabel('Freqency')
```

## Relationship between gender of the user and trip duration

> Others are more likely to spend more time in the trip

```python
sb.barplot(x='member_gender', y='duration_sec', data=df);
```

## the relation between trip duration with user type and gender 
> Here we can find that male subscribers spend more time on trips than male customers which is the same with all of the genders

```python
# Visuallizing the relation between trip duration with user type and gender 
plt.figure(figsize =[10, 10])
sb.stripplot(data =df, x="user_type", y="duration_sec", hue = 'member_gender', jitter= 0.35, dodge =True)
# Naming the title 
plt.title("The relation between Trip Duration with user type and Gender")
# NAming x axis 
plt.xlabel("User Type")
# Naming y axis 
plt.ylabel("Trip Duration in seconds");
```
